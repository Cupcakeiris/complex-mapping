```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def mapping(xmin, xmax, ymin, ymax, nb_grid, nb_points, func):
    xv, yv = np.meshgrid(np.linspace(xmin, xmax, nb_grid), np.linspace(ymin, ymax, nb_points))
    # linspace is like in matlab
    # meshgrid is used to create a grid in the matrix form
    xv = np.transpose(xv)   # we transpose, so we get transposed vectors to draw vertical lines
    yv = np.transpose(yv)

    zv = func(xv + 1j*yv)   # here's mapping which we apply to get new grid's vertical vectors, j = sqrt(-1)
    uv = np.real(zv)
    vv = np.imag(zv)

    xh, yh = np.meshgrid(np.linspace(xmin, xmax, nb_points), np.linspace(ymin, ymax, nb_grid))

    zh = func(xh + 1j*yh)   # here's mapping which we apply to get new grid's horizontal vectors
    uh = np.real(zh)
    vh = np.imag(zh)


    ax = plt.subplot(121)       # draws graph to the left (1 row, 2 columns, 1st column)
    for i in range(len(yv)):    # draws each point of grid's vertical and horizontal vectors
        ax.plot(xv[i], yv[i], 'b-', lw=1)   
        ax.plot(xh[i], yh[i], 'r-', lw=1)

    ax2 = plt.subplot(122)      # draws graph to the right (1 row, 2 columns, 2nd column)
    for i in range(len(vv)):
        ax2.plot(uv[i], vv[i], 'b-', lw=1)
        ax2.plot(uh[i], vh[i], 'r-', lw=1)

    plt.show()
```

```python
nb_grid = 15           # number of grid (cells count)
nb_points = 30         # number of points to be drawn equally

xmin, xmax, ymin, ymax = -1, 1, -1, 1   # range of grid
```

```python
mapping(xmin, xmax, ymin, ymax, nb_grid, nb_points, lambda x: x**2)
```

```python
mapping(xmin, xmax, ymin, ymax, nb_grid, nb_points, lambda x: x**3)
```

```python
mapping(xmin, xmax, ymin, ymax, nb_grid, nb_points, lambda x: x**12)
# now again to the left
```
