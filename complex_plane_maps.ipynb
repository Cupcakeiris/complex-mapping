```python
nb_grid = 15           # number of grid (cells count)
nb_points = 30         # number of points to be drawn equally

xmin, xmax, ymin, ymax = -1, 1, -1, 1   # range of grid
```

```python
a = 1 + 1j
b = -1 + 1j
c = -1 - 1j
d = 1 - 1j
```

```python
def mapp_dot(xmin, xmax, ymin, ymax, nb_grid, nb_points, func, a,b,c,d):
    xv, yv = np.meshgrid(np.linspace(xmin, xmax, nb_grid), np.linspace(ymin, ymax, nb_points))
    xv = np.transpose(xv)
    yv = np.transpose(yv)

    zv = func(xv + 1j*yv)
    uv = np.real(zv)
    vv = np.imag(zv)

    xh, yh = np.meshgrid(np.linspace(xmin, xmax, nb_points), np.linspace(ymin, ymax, nb_grid))

    zh = func(xh + 1j*yh)
    uh = np.real(zh)
    vh = np.imag(zh)


    ax = plt.subplot(121)
    for i in range(len(yv)):
        ax.plot(xv[i], yv[i], 'b-', lw=0.5)   
        ax.plot(xh[i], yh[i], 'r-', lw=0.5)

    ax.plot(np.real(a), np.imag(a), 'b.',  markersize=15)
    ax.plot(np.real(b), np.imag(b), 'g.',  markersize=15)
    ax.plot(np.real(c), np.imag(c), 'r.',  markersize=15)
    ax.plot(np.real(d), np.imag(d), 'y.',  markersize=15)

    ax2 = plt.subplot(122)
    for i in range(len(vv)):
        ax2.plot(uv[i], vv[i], 'b-', lw=0.5)
        ax2.plot(uh[i], vh[i], 'r-', lw=0.5)

    ax2.plot(np.real(func(a)), np.imag(func(a)), 'b.', markersize=15)
    ax2.plot(np.real(func(b)), np.imag(func(b)), 'g.',  markersize=15)
    ax2.plot(np.real(func(c)), np.imag(func(c)), 'r.',  markersize=15)
    ax2.plot(np.real(func(d)), np.imag(func(d)), 'y.',  markersize=15)

    plt.show()
```

```python

mapp_dot(xmin, xmax, ymin, ymax, nb_grid, nb_points, lambda x: x**2, a, b, c, d)
```
